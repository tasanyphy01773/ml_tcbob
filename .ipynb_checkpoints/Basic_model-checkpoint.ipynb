```python
#select features and target
X = dataset.iloc[:,:17].values
Y = np.column_stack((dataset['lon+6'], dataset['lat+6']))
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='constant')
X = imputer.fit_transform(X)
Y = imputer.fit_transform(Y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
print(X)  # In a certain range
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=10)

rfr = RandomForestRegressor(n_estimators=100,random_state=10)
rfr.fit(X_train,Y_train)

# Evaluate the model on the test set
Y_pred = rfr.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)

print(f"Random Forest Regressor model mean squared error: {mse}")
```

```python
lon_test = Y_test[:,0]
lat_test = Y_test[:,1]

lon_pred = Y_pred[:,0]
lat_pred = Y_pred[:,1]

random_dataset = pd.DataFrame()
random_dataset['lon_test'] = lon_test
random_dataset['lat_test'] = lat_test

random_dataset['lon_pred'] = lon_pred
random_dataset['lat_pred'] = lat_pred
random_dataset['deviation'] = np.sqrt(((lon_test-lon_pred)**2)+ ((lat_test-lat_pred)**2))

#average error in deviation
print("  ")
print("The average error for the deviation is "+ str(random_dataset['deviation'].mean()))
print("  ")
#plt.plot(lon_test, lat_test)
random_dataset
```

```python
plt.figure()
plt.scatter(lon_test, lat_test)
plt.scatter(lon_pred, lat_pred)
plt.xlabel('lon')
plt.ylabel('lat')

plt.legend(['Test', 'Prediction'], loc='upper left')
```
